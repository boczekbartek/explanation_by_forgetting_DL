```python
import matplotlib.pyplot as plt
import pandas as pd
from glob import glob
from matplotlib import rc
import os 
os.environ['PATH'] = f"/Library/TeX/texbin/:{os.environ['PATH']}"
plt.style.use('lol.mplstyle')

# plt.rcParams.update({'font.size': 14})

# plt.rcParams['text.latex.preamble'] = [r'\usepackage{mathptmx}']  # load times roman font
# plt.rcParams['font.family'] = 'serif'  # use serif font as default
# plt.rcParams['text.usetex'] = True  # enable LaTeX rendering globally
plt.rcParams['axes.facecolor'] = 'white'
## for Palatino and other serif fonts use:
# rc('font',**{'family':'serif','serif':['Palatino']})
```

```python
for n in (10, 50, 100):
    plt.figure()
    for method,m in (('random', 'random'), ('most_freq','most frequent')):
        f = f'cell_ontology/workdir_{n}_{method}.stats.tsv'
        if not os.path.isfile(f):
            continue
        vals = pd.read_csv(f, sep='\t')['Length'].values
        plt.plot(vals, label=m)
    plt.legend()
    plt.xlabel('Step')
    plt.ylabel('Explanation length')
    plt.title(f'{n} symbols forgotten in each iteration.')
    plt.savefig(f'cell_ontology/cell_forget{n}.pdf', dpi=300)
```
